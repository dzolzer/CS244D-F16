### Scores
Reading and plotting scores with numpy and matplotlib

```python
%matplotlib inline
```

```python
import numpy as np
from matplotlib import pyplot as pl
```

```python
def read_scores(file_name):
    dtypes = np.dtype({'names' : ('student', 'score'),
                       'formats' : [np.int, np.int]})
    
    scores = np.loadtxt(file_name, delimiter=',', dtype=dtypes)
    return scores
```

```python
def plot_scores(student,score):
    fig = pl.figure()
    pl.title("Scores")
    pl.ylabel('Scores')
    pl.xlabel("Students")
    pl.plot(student, score)
    
    slope, intercept = np.polyfit(student, score, 1)
    ideal_score = intercept + (slope * student)
    
    pl.plot(student, ideal_score, color='red', linestyle='--')
    
    return fig
```

```python
# this still needs work. 
def plot_hist(student, score):
    fig = pl.figure()
    pl.title('Score Distribution')
    
    bins = np.arange(1,5+2)
    pl.hist(score, bins=bins, label=score)
    pl.legend()
    return fig
```

```python
data = read_scores('scores.csv')
student = data['student']
score = data['score']
fig = plot_scores(student, score)
fig.savefig('scores.png')
fig = plot_hist(student, score)
fig.savefig('score_histogram.png')
```
